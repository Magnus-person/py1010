```python
"""
Oppgave 1.
Ett program som starter med 
alder = int(input('Hvilket år er du født? ') )
og skal regne ut hvor gammel personen blir i 2024.
Jeg antar at det er ønsket at det regner ut alderen i år og jeg har derfor tatt med begge deler. 

Created: 7.11.25

author: Magnus Slettjord, magnusslettjord@hotmail.com
"""
import datetime
def aar_i_aar():
    return datetime.date.today().year  # Bruker datetime modulen for å definere året i år
    
alder = int(input('Hvilket år er du født? ') )  # etterspør alderen til personen
print('Du ble', 2024-alder, 'år i 2024')  # regner ut og skriver til skjerm personens alder i 2024
print ('I år blir du', aar_i_aar()-alder, 'år')  # regner ut og skriver til skjerm personens alder i år
```

```python
"""
Oppgave 2.
Lag ett program som tar inn antall elever fra konsollen ved 
antall_elever = int(input('Skriv inn antall elever:' )) 
Programmet antar at hver elev spiser 1/4 pizza,
og regner så ut hvor mange pizzaer som skal handles inn til festen og skriver svaret til skjerm. 


Created: 7.11.25

author: Magnus Slettjord, magnusslettjord@hotmail.com
"""
import math

antall_elever = int(input('Skriv inn antall elever:' ))  # Den obligatoriske delen av oppgaven
antall_pizza = math.ceil(antall_elever/4)  # antallet elever deles på 4 og math.ceil runder opp resultatet. 

if antall_pizza == 1:  # Sjekker om en pizza holder og skriver i såfall det
    print('Det må kjøpes inn 1 pizza')
else :
    print ('Det må kjøpes inn', antall_pizza, 'pizzaer')  # Skriver ut hvis antallet er mer enn 1
```

```python
"""
Oppgave 3.
Et program som regner grader til radianer

Created: 8.11.25

author: Magnus Slettjord, magnusslettjord@hotmail.com
"""

import numpy as np  
import sympy as sp

v_grad = float(input('Skriv inn gradtallet:' ))  # Definerer grader basert på input
v_rad = v_grad*np.pi/180  # Regner om grader til radianer
v_rad_symbolsk = sp.simplify(sp.Rational(v_grad) * sp.pi /180)  # Bruker sympy for å regne ut radianer uttrykt som brøk med π

print(f'{v_grad} grader = {v_rad:.4f} eller {v_rad_symbolsk} radianer')  # Skriver ut antall grader, antall radianer med 4 desimaler og uttrykt som en brøk av pi
```

```python
"""
Oppgave 5
Program med en funksjon som tar a og b som inn-argumenter og regner ut arealet 
og ytre omkrets til en figur satt sammen av en rettviknklet trekant og en halvsirkel 
hvor a er både den lille kateten til trekanten og diameteren til halvsirkelen. 
Funksjonen returnerer arealet og ytre omkrets og skriver resultatet til skjerm med passende tekst. 

Created: 8.11.25

author: Magnus Slettjord, magnusslettjord@hotmail.com
"""
import math 

# Definerer en funksjon som regner ut omkretsen og arealet av figuren. 
def figur(a, b):
    omkrets_lokal = (2*math.pi*a/2) + b + (math.sqrt(a**2 + b**2))
    areal_lokal = (math.pi*(a/2)*(a/2)/2)+(a*b/2)
    
    return (omkrets_lokal, areal_lokal) 

# Etterspør variablene som brukes til å regne ut størrelsen på figuren
a = float(input('Skriv lengden på a:'))
b = float(input('Skriv lengden på b:'))


omkrets_global, areal_global = figur(a, b) 
print(f"Omkretsen på figuren er {omkrets_global:.2f} og arealet er {areal_global:.2f}")
```
